```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob, Word

%matplotlib inline
```

```python
# define a function that accepts text and returns the polarity using TextBlob sentiment analyzer
def detect_sentiment(text):
    return TextBlob(str(text).decode('utf-8')).sentiment.polarity
```

```python
df = pd.read_csv('amazon_food.csv', index_col=0)
```

```python
df.head()
```

```python
# create a new DataFrame column for sentiment (WARNING: SLOW!)
df['sentiment_polarity'] = df['reviewText'].apply(detect_sentiment)
```

```python
df.sentiment_polarity.head()
```

```python
# box plot of sentiment grouped by overall rating
df.boxplot(column='sentiment_polarity', by='overall', fontsize=18, figsize=(15,10))
plt.title('Predicted Sentiment vs Actual Rating', fontsize=18)
plt.xlabel('Actual Overall Rating', fontsize = 18)
plt.ylabel('Predicted Sentiment Polarity', fontsize=18)
plt.show()
```
