```python
import time

start = time.time()
                                
import sys
import os

searchQuery = 'food'  # this is what we're searching for
maxTweets = 100000 # Some arbitrary large number
tweetsPerQry = 100  # this is the max the API permits

# If results from a specific ID onwards are reqd, set since_id to that ID.
# else default to no lower limit, go as far back as API allows
sinceId = None

# If results only below a specific ID are, set max_id to that ID.
# else default to no upper limit, start from the most recent tweet matching the search query.
max_id = 917635767819583488

tweets = []
tweetCount = 0
print("Downloading max {0} tweets".format(maxTweets))

while tweetCount < maxTweets:
    try:
        if (max_id <= 0):
            if (not sinceId):
                new_tweets = api.search(q=searchQuery, count=tweetsPerQry)
            else:
                new_tweets = api.search(q=searchQuery, count=tweetsPerQry,
                                        since_id=sinceId)
        else:
            if (not sinceId):
                new_tweets = api.search(q=searchQuery, count=tweetsPerQry,
                                        max_id=str(max_id - 1))
            else:
                new_tweets = api.search(q=searchQuery, count=tweetsPerQry,
                                        max_id=str(max_id - 1),
                                        since_id=sinceId)
        if not new_tweets:
            print("No more tweets found")
            break
            
        for tweet in new_tweets:
            tweets.append(tweet)

        tweetCount += len(new_tweets)
        print("Downloaded {0} tweets".format(tweetCount))
        
        max_id = new_tweets[-1].id
    
    except tweepy.TweepError as e:
        # Just exit if any error
        print("some error : " + str(e))
        break

end = time.time()
exe_time = end - start
print 'Time taken :',(exe_time),' seconds'
```

```python
start = time.time()

def no_duplicates(results):
    # to remove duplicate tweets that are not retweets by comparing the first 25 characters of the tweet

    no_dupli = []
    # initialise empty list to contain duplicate tweets once only 
    dup_once = []
    # initialise dup_once ist to contain truncated tweets for comparison
    dup_once_trunc = []
    res_trunc = [x[:25] for x in results]

    for i in range(len(results)):
        # remove truncated item to see if there are other duplicates in the list
        del res_trunc[i]
        if results[i][:25] not in res_trunc:
            no_dupli.append(results[i])
        else:
            # check and input duplicates once only
            if results[i][:25] not in dup_once_trunc:
                    dup_once.append(results[i])
                    dup_once_trunc.append(results[i][:25])
        res_trunc = [x[:25] for x in results]

    no_dupli.extend(dup_once)
            
    return no_dupli

no_dup = no_duplicates(results)

print len(no_dup)

end = time.time()
exe_time = end - start
print 'Time taken :',(exe_time),' seconds'
```

```python
# parse html elements
import HTMLParser
html_parser = HTMLParser.HTMLParser()

no_html = []
for i in no_dup:
    i = html_parser.unescape(i)
    no_html.append(i)
    
print len(no_html)
```

```python
# remove urls at the end of the tweet
url_re = 'https:\/\/t\.co\/([a-zA-Z\d]+)'
    
no_url = []
for i in no_html:
    u = re.finditer(url_re , i)
    if u:
        for a in u:
            i = i.replace(a.group(), '')
    # remove newlines including those created when url removed at end of string
    i = i.replace('\n', '')
    no_url.append(i)
    
print len(no_url)
```

```python
# remove any screen names encountered
no_scrname = []
scrname_re = '@([a-zA-Z\d]*)'
for i in no_url:
    u = re.finditer(scrname_re , i)
    if u:
        for a in u:
            i = i.replace(a.group(), '')
    no_scrname.append(i)    
```

```python
# remove all hashtags
no_hashtag = []
hashtag_re = '#([a-z]+)'

for i in no_scrname:
    u = re.finditer(hashtag_re , i)
    if u:
        for a in u:
            i = i.replace(a.group(), '')
    no_hashtag.append(i)
```

```python
no_spec = []
spec_char = ['$', '(', ')','{','}','~','@','#','%','^','&','*',':','|','<',';','[',']']

for i in no_hashtag:
    for char in spec_char:
        i = i.replace(char, '')
    no_spec.append(i)
```

```python
no_num = []
num_re = '\d'

for i in no_spec:
    u = re.findall(num_re, i)
    if u:
        for a in u:
            i = i.replace(a , '')
    no_num.append(i)
```

```python
# get all a-z characters only and remove emojis, punctuations and special characters
char_only = []
not_char_re = '([^a-z\s]+)'

for i in no_num:
    u = re.findall(not_char_re , i)
    if u:
        for a in u:
            i = i.replace(a , '')
    char_only.append(i)
```

```python
tweets = pd.DataFrame(char_only, columns = ['Tweets'])
tweets.tail(10)
```
