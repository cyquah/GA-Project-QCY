```python
import pandas as pd
from textblob import TextBlob, Word
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
df = pd.read_csv('tamchiak.csv', index_col=0)
```

```python
df.head()
```

```python
tweets = df.Tweets
print tweets[0]
```

```python
import time

start = time.time()

spell_check = []
for tweet in tweets:
    tweet = str(tweet).decode('utf-8')
    t = TextBlob(tweet).correct()
    spell_check.append(t)
    
end = time.time()
exe_time = end - start
print 'Time taken :',(exe_time),' seconds'
```

```python
start = time.time()

lemma_check =[]
wnl = WordNetLemmatizer()
for tweet in tweets:
    tweet = str(tweet).decode('utf-8')
    t = " ".join([wnl.lemmatize(i) for i in tweet.split()])
    lemma_check.append(t) 
    
end = time.time()
exe_time = end - start
print 'Time taken :',(exe_time),' seconds'
```

```python
word_count = []
for lemma in lemma_check:
    count = len(lemma.split())
    word_count.append(count)
```

```python
df['Tw_processed'] = lemma_check
```

```python
df['Word_count'] = word_count
sum(df['Word_count'])
```

```python
print 'Total number of records :', len(df)
print 'Total number of words :', sum(df['Word_count'])
print 'Average no. of words per record :', sum(df['Word_count'])/len(df)
```
